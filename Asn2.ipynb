```python
#Connect to SQL data file and run query
import sqlite3 as sql
import pandas as pd
connect = sql.Connection("dj.db")
cursor = connect.cursor()
cursor.execute("SELECT id, ticker, name FROM tickers")
stonks = cursor.fetchall()
```

```python
stonks
```

```python
for values in stonks:
    print(values[0] + values[1]+ values[2])
```

```python
#Turn data output from SQL file into pandas DF
stonksDataFrame = pd.DataFrame(stonks)
stonksDataFrame.set_index(1, inplace=True)
stonksDataFrame.columns = ['id', 'name']
stonksDataFrame.index.name = 'Ticker'
stonksDataFrame
```

```python
#Use query to find the earliest and latest date of the data in table specified in query
cursor.execute("SELECT min(date) as first, max(date) as last FROM daily_measures")
date = cursor.fetchall()
date
```

```python
#USe query to find specific data on table from data file then turn output into pandas DF
cursor.execute("SELECT date, closing FROM daily_measures WHERE ticker_id=3 ORDER BY date")
part3 = cursor.fetchall()
applData = pd.DataFrame(part3)
applData.columns = ['Date', 'Close Value']
applData
```

```python
#Find max value from previous query using created DF
maxValue = max(applData['Close Value'])
applData.loc[applData['Close Value'] == maxValue]
```

```python
#Append new columns to DF that displays daily changes of data from query
applData["Daily Return"] = applData['Close Value'].pct_change()
applData
```

```python
import numpy as np
```

```python
#USing numpy to identify max value in DF
maxChangePos = np.nanmax(applData['Daily Return'])
maxChangeNeg = np.nanmin(applData['Daily Return'])
if maxChangePos > abs(maxChangeNeg):
    maxChange = maxChangePos
else:
    maxChange = maxChangeNeg
applData.loc[applData['Daily Return'] == maxChange]
```
